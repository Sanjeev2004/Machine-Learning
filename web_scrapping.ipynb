```python
import requests

url = "https://en.wikipedia.org/wiki/Kama_Sutra"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

print(response.text)  # HTML content of the page
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")
titles = soup.find_all("h2")  # Extract all h2 elements

for title in titles:
    print(title.text.strip())  # Print the text inside h2
```

```python
from selenium import webdriver

driver = webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/Kama_Sutra")

page_content = driver.page_source  # Get HTML after JavaScript execution
driver.quit()
```
