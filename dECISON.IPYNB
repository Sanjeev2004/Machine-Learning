```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
data = pd.read_csv('play_tennis.csv')
print(data.head())
```

```python
X = data.drop(columns=['play'],axis=1)  # Features (Outlook, Temperature, Humidity, Wind)
y = data['play']  # Target variable (PlayTennis)
```

```python

X = pd.get_dummies(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
clf_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf_entropy.fit(X_train, y_train)
```

```python
y_pred_entropy = clf_entropy.predict(X_test)
print("Accuracy using Entropy (Information Gain):", metrics.accuracy_score(y_test, y_pred_entropy))
```

```python
# Decision Tree using Gini Index
clf_gini = DecisionTreeClassifier(criterion='gini', random_state=42)
clf_gini.fit(X_train, y_train)
```

```python
# Predictions and accuracy using Gini
y_pred_gini = clf_gini.predict(X_test)
print("Accuracy using Gini Index:", metrics.accuracy_score(y_test, y_pred_gini))
```

```python
# Optional: Visualize the decision tree
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
tree.plot_tree(clf_entropy, feature_names=X.columns, class_names=True, filled=True)
plt.show()
```
